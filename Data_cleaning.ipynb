```python
# SQL CONNECTION AND UPLOAD

from sqlalchemy import create_engine
import mysql.connector

# --------- CONFIG ---------
db_name = 'agri_data'
table_name = 'agriculture'

mysql_config = {
    'host': 'localhost',
    'user': 'root',       # XAMPP default
    'password': ''        # Set your MySQL password if you have one
}

# --------- STEP 1: CREATE DATABASE IF NOT EXISTS ---------
mydb = mysql.connector.connect(**mysql_config)
mydb.autocommit = True
mycursor = mydb.cursor()
mycursor.execute(f"CREATE DATABASE IF NOT EXISTS {db_name}")
print(f"Database '{db_name}' checked/created.")
mydb.commit()
```

```python
# --------- STEP 2: CHECK CSV COLUMNS---------
print(f"CSV loaded. Columns: {list(df.columns)}")
```

```python

# --------- STEP 3: CONNECT TO MYSQL USING SQLALCHEMY ---------
connection_str = f"mysql+mysqlconnector://{mysql_config['user']}:{mysql_config['password']}@{mysql_config['host']}:3306/{db_name}"
engine = create_engine(connection_str)
```

```python

# --------- STEP 4: CREATE/REPLACE TABLE AND UPLOAD DATA ---------
df.to_sql(name=table_name, con=engine, if_exists='replace', index=False)
print(f"Data uploaded to '{table_name}' table in '{db_name}' database successfully!")
```
